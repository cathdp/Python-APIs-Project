```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import citipy

from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# cities = ["New York","asdsfa","London"]
cloudiness = []
country = []
date = []
humidity = []
temp = []
lat = []
lng = []
wind =[]
record_count= 1
set_count=1

url="http://api.openweathermap.org/data/2.5/weather?"
units="imperial"
key =weather_api_key

query_url=f"{url}appid={key}&units={units}&q="

print('Beginning Data Retrieval')     
print('-----------------------------')

for i,city in enumerate(cities):
    response= requests.get(query_url+city).json()
    
    if (i >= 50 and i % 50 == 0):
        set_count+=1
    
    print(f'Processing Record {record_count} of Set {set_count} | {city}')
    record_count+=1
   
    try:
        id=response['id']
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        date.append(response['dt'])
        country.append(response['sys']['country'])
    except (KeyError, IndexError):
        lat.append(None)
        lng.append(None)
        temp.append(None)
        humidity.append(None)
        cloudiness.append(None)
        wind.append(None)
        date.append(None)
        country.append(None)    
        print('City not found. Skipping...')
        pass
print('-----------------------------')
print('Data Retrieval Complete      ')
print('-----------------------------')
```

* Export the city data into a .csv.

```python
weather_dict = {
    "City":cities,
    "Cloudiness":cloudiness,
    "Country":country,
    "Date":date,
    "Humidity":humidity,
    "Lat":lat,
    "Lng":lng,
    "Temp":temp,
    "Wind Speed":wind}

weather_data = pd.DataFrame(weather_dict)

weather_data = weather_data.dropna()
weather_data.head()
    
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
Northern_data = weather_data.loc[(weather_data["Lat"]>0),:]
Southern_data = weather_data.loc[(weather_data["Lat"]<0),:]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Lin_Reg(Northern_data["Lat"],Northern_data["Temp"])
x_values= Northern_data ['Lat']
y_values= Northern_data ['Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Latitude Linear Regression')
plt.ylabel('Max Temp')
print(f"The r-squared is: {rvalue}")
plt.show()
```
